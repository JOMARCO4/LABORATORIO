```python
# 1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. 
# Los espacios no deben ser considerados
```

```python
def frecuencias_letras(texto: str) -> dict[str, int]:
    """
    Calcula la frecuencia de cada letra en una cadena de texto.

    Los espacios no se consideran y las mayúsculas se convierten a minúsculas.

    Parámetros:
        texto (str): Cadena de texto de entrada.

    Retorna:
        dict[str, int]: Diccionario con cada letra como clave
                        y su frecuencia como valor.
    """
    frecuencias: dict[str, int] = {}

    for caracter in texto.lower():
        if caracter != " ":  # Ignoramos espacios
            frecuencias[caracter] = frecuencias.get(caracter, 0) + 1

    return frecuencias


# Ejemplo de uso
print(frecuencias_letras("Esto es Python"))
# Salida esperada: {'e': 2, 's': 2, 't': 2, 'o': 2, 'p': 1, 'y': 1, 'h': 1, 'n': 1}
```

```python
# 2. Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función map()
```

```python
from typing import List

def duplicar_lista(numeros: List[int]) -> List[int]:
    """
    Devuelve una nueva lista con el doble de cada valor de la lista original.

    Parámetros:
        numeros (List[int]): Lista de números de entrada.

    Retorna:
        List[int]: Nueva lista con cada valor duplicado.
    """
    return list(map(lambda x: x * 2, numeros))


# Ejemplo de uso
print(duplicar_lista([4, 6, 3, 7, 10]))  # Salida: [8, 12, 6, 14, 20]
```

```python
# 3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros.
# La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.
```

```python
from typing import List

def filtrar_palabras_con_objetivo(palabras: List[str], objetivo: str) -> List[str]:
    """
    Devuelve una lista con todas las palabras de la lista original que contienen la palabra objetivo.

    Args:
        palabras (List[str]): Lista de palabras a filtrar.
        objetivo (str): Subcadena que debe estar presente en las palabras.

    Returns:
        List[str]: Lista de palabras que contienen la subcadena objetivo.
    """
    return [palabra for palabra in palabras if objetivo in palabra]


# Ejemplo de uso
lista_palabras = ["mercedes", "audi", "bmw", "seat", "skoda"]
palabra_objetivo = "s"

resultado = filtrar_palabras_con_objetivo(lista_palabras, palabra_objetivo)
print(resultado)  # Salida: ['mercedes', 'seat', 'skoda']
```

```python
# 4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función map()
```

```python
from typing import List

def calcular_diferencias(lista1: List[float], lista2: List[float]) -> List[float]:
    """
    Calcula la diferencia entre los elementos de dos listas de igual longitud.

    Args:
        lista1 (List[float]): Primera lista de números.
        lista2 (List[float]): Segunda lista de números.

    Returns:
        List[float]: Lista con las diferencias elemento a elemento (lista1[i] - lista2[i]).
    
    Raises:
        ValueError: Si las listas no tienen la misma longitud.
    """
    if len(lista1) != len(lista2):
        raise ValueError("Las listas deben tener la misma longitud.")

    return list(map(lambda x, y: x - y, lista1, lista2))


# Ejemplo de uso
numeros1 = [20, 30, 50, 60]
numeros2 = [1, 5, 10, 50]

resultado = calcular_diferencias(numeros1, numeros2)
print(resultado)  # Salida: [19, 25, 40, 10]
```

```python
# 5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. 
# La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota aprobado. 
# Si es así, el estado será "aprobado", de lo contrario, será "suspenso". 
# La función debe devolver una tupla que contenga la media y el estado.
```

```python
from typing import List, Tuple

def evaluar_media(numeros: List[float], nota_aprobado: float = 5.0) -> Tuple[float, str]:
    """
    Calcula la media de una lista de números y determina si está aprobado según la nota de corte.

    Args:
        numeros (List[float]): Lista de números a evaluar.
        nota_aprobado (float, opcional): Nota mínima para aprobar. Por defecto es 5.0.

    Returns:
        Tuple[float, str]: Una tupla con la media y el estado ('aprobado' o 'suspenso').

    Raises:
        ValueError: Si la lista de números está vacía.
    """
    if not numeros:
        raise ValueError("La lista de números no puede estar vacía.")
    
    media = sum(numeros) / len(numeros)
    estado = "aprobado" if media >= nota_aprobado else "suspenso"
    return media, estado


# Ejemplo de uso
notas = [6.5, 8.1, 9, 5]
resultado = evaluar_media(notas)
print(resultado)  # Salida: (7.15, 'aprobado')
```

```python
# 6. Escribe una función que calcule el factorial de un número de manera recursiva.
```

```python
def factorial(n: int) -> int:
    """
    Calcula el factorial de un número entero no negativo de manera recursiva.

    Args:
        n (int): Número entero no negativo.

    Returns:
        int: Factorial de n.

    Raises:
        ValueError: Si n es negativo.
    """
    if n < 0:
        raise ValueError("El número debe ser no negativo.")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


# Ejemplo de uso
numero = 6
resultado = factorial(numero)
print(resultado)  # Salida: 720
```

```python
# 7. Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función map()
```

```python
from typing import List, Tuple

def tuplas_a_strings(lista_tuplas: List[Tuple]) -> List[str]:
    """
    Convierte una lista de tuplas en una lista de strings.

    Args:
        lista_tuplas (List[Tuple]): Lista de tuplas a convertir.

    Returns:
        List[str]: Lista de strings donde cada string representa una tupla.
    """
    # Usamos map para aplicar str() a cada tupla
    return list(map(str, lista_tuplas))


# Ejemplo de uso
lista = [(45, 12), (9, 11), ("z", "t")]
resultado = tuplas_a_strings(lista)
print(resultado)  # Salida: ['(45, 12)', '(9, 11)', "('z', 't')"]
```

```python
# 13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. 
# Las letras no pueden estar repetidas .Usa la función map()
```

```python
from typing import List, Tuple

def letras_may_min(caracteres: str) -> List[Tuple[str, str]]:
    """
    Devuelve una lista de tuplas con cada letra en mayúscula y minúscula,
    sin repetir letras.

    Args:
        caracteres (str): Conjunto de caracteres de entrada.

    Returns:
        List[Tuple[str, str]]: Lista de tuplas (mayúscula, minúscula).
    """
    # Convertir a set para eliminar duplicados y filtrar solo letras
    letras_unicas = {c.lower() for c in caracteres if c.isalpha()}
    # Crear tuplas (mayúscula, minúscula) usando map
    return list(map(lambda c: (c.upper(), c.lower()), letras_unicas))

# Ejemplo de uso
entrada = "aaAbBcCcccDada"
resultado = letras_may_min(entrada)
print(resultado)  # Output: [('A', 'a'), ('B', 'b'), ('C', 'c'), ('D', 'd')] (el orden puede variar)
```

```python
# 14. Crea una función que retorne las palabras de una lista de palabras que comience con una letra en especifico. Usa la función filter()
```

```python
from typing import List

def filtrar_por_letra(palabras: List[str], letra: str) -> List[str]:
    """
    Retorna las palabras que comienzan con una letra específica.

    Args:
        palabras (List[str]): Lista de palabras.
        letra (str): Letra con la que deben comenzar las palabras.

    Returns:
        List[str]: Lista de palabras que comienzan con la letra dada.
    """
    letra = letra.lower()
    return list(filter(lambda palabra: palabra.lower().startswith(letra), palabras))

# Ejemplo de uso
lista_palabras = ["Plátano", "Manzana", "pera", "piña", "Melón", "banana", "Mango"]
resultado = filtrar_por_letra(lista_palabras, "p")
print(resultado)  # Output: ['Plátano', 'pera', 'piña']
```

```python
from typing import List

def palabras_mas_largas(texto: str, n: int) -> List[str]:
    """
    Devuelve una lista con las palabras de la cadena que tienen más de n caracteres.

    Args:
        texto (str): Cadena de texto.
        n (int): Longitud mínima de las palabras a incluir.

    Returns:
        List[str]: Lista de palabras con longitud mayor que n.
    """
    palabras = texto.split()
    return list(filter(lambda palabra: len(palabra) > n, palabras))

# Ejemplo de uso
frase = "Python es un lenguaje de programación muy potente"
resultado = palabras_mas_largas(frase, 5)
print(resultado)  # Output: ['lenguaje', 'programación', 'potente']
```

```python
# 17. Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2] corresponde al número quinientos setenta y dos (572).
# Usa la función reduce()
```

```python
from functools import reduce
from typing import List

def lista_a_numero(digitos: List[int]) -> int:
    """
    Convierte una lista de dígitos en el número correspondiente.

    Args:
        digitos (List[int]): Lista de dígitos (0-9).

    Returns:
        int: Número formado por los dígitos.
    """
    if not digitos:
        raise ValueError("La lista de dígitos no puede estar vacía.")
    if not all(0 <= d <= 9 for d in digitos):
        raise ValueError("Todos los elementos deben ser dígitos entre 0 y 9.")
    
    return reduce(lambda total, d: total * 10 + d, digitos)

# Ejemplo de uso
digitos = [5, 7, 2]
resultado = lista_a_numero(digitos)
print(resultado)  # Output: 572
```

```python
from functools import reduce
from typing import List

def concatenar_palabras(palabras: List[str]) -> str:
    """
    Concatena todas las palabras de una lista en una sola cadena.

    Args:
        palabras (List[str]): Lista de palabras.

    Returns:
        str: Cadena resultante de la concatenación de todas las palabras.
    """
    if not palabras:
        return ""  # Retorna cadena vacía si la lista está vacía
    
    return reduce(lambda x, y: x + y, palabras)

# Ejemplo de uso
lista = ["Hola", " ", "queridos", " ", "amigos", "!"]
resultado = concatenar_palabras(lista)
print(resultado)  # Output: "Hola  queridos  amigos!"
```

```python
# 28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.
```

```python
from typing import List, Any, Optional

def primer_duplicado(lista: List[Any]) -> Optional[Any]:
    """
    Devuelve el primer elemento duplicado en la lista, si existe.

    Args:
        lista (List[Any]): Lista de elementos.

    Returns:
        Optional[Any]: Primer elemento duplicado, o None si no hay duplicados.
    """
    vistos = set()
    for elemento in lista:
        if elemento in vistos:
            return elemento
        vistos.add(elemento)
    return None

# Ejemplo de uso
lista = [1, 2, 3, 4, 2, 5, 3]
resultado = primer_duplicado(lista)
print(resultado)  # Output: 2
```

```python
from typing import Any

def enmascarar_variable(valor: Any) -> str:
    """
    Convierte cualquier valor en cadena y enmascara todos los caracteres
    con '#' excepto los últimos cuatro.

    Args:
        valor (Any): Valor a enmascarar.

    Returns:
        str: Cadena enmascarada con los últimos cuatro caracteres visibles.
    """
    texto = str(valor)
    longitud = len(texto)
    if longitud <= 4:
        return texto  # Si tiene 4 o menos caracteres, no se enmascara
    return '#' * (longitud - 4) + texto[-4:]

# Ejemplo de uso
numero = 1234567890
resultado = enmascarar_variable(numero)
print(resultado)  # Output: "######7890"

texto = "abcdef"
resultado_texto = enmascarar_variable(texto)
print(resultado_texto)  # Output: "##cdef"
```

```python
from typing import Optional

def son_anagramas(palabra1: str, palabra2: str) -> bool:
    """
    Determina si dos palabras son anagramas.

    Args:
        palabra1 (str): Primera palabra.
        palabra2 (str): Segunda palabra.

    Returns:
        bool: True si son anagramas, False en caso contrario.
    """
    # Eliminamos espacios y convertimos a minúsculas para una comparación más precisa
    palabra1 = palabra1.replace(" ", "").lower()
    palabra2 = palabra2.replace(" ", "").lower()
    
    # Comparamos las palabras ordenando sus caracteres
    return sorted(palabra1) == sorted(palabra2)

# Ejemplo de uso
resultado1 = son_anagramas("amor", "roma")
print(resultado1)  # Output: True

resultado2 = son_anagramas("hola", "adios")
print(resultado2)  # Output: False
```

```python
# 34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: crecer_tronco , nueva_rama , crecer_ramas , 
# quitar_rama e info_arbol . El objetivo es implementar estos métodos para manipular la estructura del árbol.
#      Código a seguir:
#           1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
#           2. Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
#           3. Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
#           4. Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
#           5. Implementar el método quitar_rama para eliminar una rama en una posición específica.
#           6. Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.
#       Caso de uso:
#           1. Crear un árbol.
#           2. Hacer crecer el tronco del árbol una unidad.
#           3. Añadir una nueva rama al árbol.
#           4. Hacer crecer todas las ramas del árbol una unidad.
#           5. Añadir dos nuevas ramas al árbol.
#           6. Retirar la rama situada en la posición 2.
#           7. Obtener información sobre el árbol.
```

```python
class Arbol:
    """
    Clase que representa un árbol genérico con un tronco y ramas.

    Atributos:
        tronco (int): Longitud del tronco.
        ramas (list[int]): Lista con la longitud de cada rama.
    """

    def __init__(self):
        """Inicializa un árbol con tronco de longitud 1 y lista vacía de ramas."""
        self.tronco = 1
        self.ramas = []

    def crecer_tronco(self):
        """Aumenta la longitud del tronco en 1 unidad."""
        self.tronco += 1

    def nueva_rama(self):
        """Agrega una nueva rama de longitud 1 a la lista de ramas."""
        self.ramas.append(1)

    def crecer_ramas(self):
        """Aumenta en 1 unidad la longitud de todas las ramas existentes."""
        self.ramas = [rama + 1 for rama in self.ramas]

    def quitar_rama(self, posicion):
        """
        Elimina una rama en una posición específica.

        Args:
            posicion (int): Índice de la rama a eliminar (0-based).

        Devuelve:
            bool: True si se eliminó correctamente, False si el índice es inválido.
        """
        if 0 <= posicion < len(self.ramas):
            del self.ramas[posicion]
            return True
        return False

    def info_arbol(self):
        """
        Devuelve información sobre el árbol.

        Returns:
            dict: Contiene longitud del tronco, número de ramas y longitudes de las ramas.
        """
        return {
            "tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitudes_ramas": self.ramas
        }


# ------------------- Caso de uso -------------------

# 1. Crear un árbol
arbol = Arbol()

# 2. Hacer crecer el tronco una unidad
arbol.crecer_tronco()

# 3. Añadir una nueva rama
arbol.nueva_rama()

# 4. Hacer crecer todas las ramas una unidad
arbol.crecer_ramas()

# 5. Añadir dos nuevas ramas
arbol.nueva_rama()
arbol.nueva_rama()

# 6. Retirar la rama situada en la posición 2 (índice 1 si se cuenta desde 0)
arbol.quitar_rama(1)

# 7. Obtener información sobre el árbol
info = arbol.info_arbol()
print(info)
# Output esperado:{'tronco': 2, 'numero_ramas': 2, 'longitudes_ramas': [2, 1]}
```

```python
# 36. Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta corriente. 
# Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y agregar dinero al saldo.
#       Código a seguir:
#           1. Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante True y False.
#           2. Implementar el método retirar_dinero para retirar dinero del saldo del usuario. Lanzará un error en caso de no poder hacerse.
#           3. Implementar el método transferir_dinero para realizar una transferencia desde otro usuario al usuario actual. Lanzará un error en caso de no poder hacerse.
#           4. Implementar el método agregar_dinero para agregar dinero al saldo del usuario.
#       Caso de uso:
#           1. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
#           2. Agregar 20 unidades de saldo de "Bob".
#           3. Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
#           4. Retirar 50 unidades de saldo a "Alicia".
```

```python
class SaldoInsuficienteError(Exception):
    """Excepción personalizada para indicar saldo insuficiente."""
    pass

class UsuarioBanco:
    """
    Clase que representa a un usuario de un banco.

    Atributos:
        nombre (str): Nombre del usuario.
        saldo (float): Saldo disponible.
        cuenta_corriente (bool): Indica si tiene cuenta corriente.
    """

    def __init__(self, nombre, saldo=0.0, cuenta_corriente=False):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):
        """
        Retira dinero del saldo del usuario.

        Args:
            cantidad (float): Cantidad a retirar.

        Raises:
            SaldoInsuficienteError: Si el saldo es insuficiente.
        """
        if cantidad <= 0:
            raise ValueError("La cantidad a retirar debe ser mayor que 0.")
        if cantidad > self.saldo:
            raise SaldoInsuficienteError(f"{self.nombre} no tiene suficiente saldo para retirar {cantidad}.")
        self.saldo -= cantidad

    def transferir_dinero(self, usuario_origen, cantidad):
        """
        Transfiere dinero desde otro usuario al usuario actual.

        Args:
            usuario_origen (UsuarioBanco): Usuario que envía el dinero.
            cantidad (float): Cantidad a transferir.

        Raises:
            SaldoInsuficienteError: Si el usuario de origen no tiene saldo suficiente.
        """
        if cantidad <= 0:
            raise ValueError("La cantidad a transferir debe ser mayor que 0.")
        if cantidad > usuario_origen.saldo:
            raise SaldoInsuficienteError(f"{usuario_origen.nombre} no tiene suficiente saldo para transferir {cantidad}.")
        usuario_origen.saldo -= cantidad
        self.saldo += cantidad

    def agregar_dinero(self, cantidad):
        """
        Agrega dinero al saldo del usuario.

        Args:
            cantidad (float): Cantidad a agregar.
        """
        if cantidad <= 0:
            raise ValueError("La cantidad a agregar debe ser mayor que 0.")
        self.saldo += cantidad

    def __str__(self):
        return f"UsuarioBanco(nombre={self.nombre}, saldo={self.saldo}, cuenta_corriente={self.cuenta_corriente})"


# ------------------- Caso de uso -------------------

# 1. Crear dos usuarios
alicia = UsuarioBanco("Alicia", saldo=100, cuenta_corriente=True)
bob = UsuarioBanco("Bob", saldo=50, cuenta_corriente=True)

# 2. Agregar 20 unidades al saldo de Bob
bob.agregar_dinero(20)
print(bob)  # Saldo esperado: 70

# 3. Hacer una transferencia de 80 unidades desde Bob a Alicia
try:
    alicia.transferir_dinero(bob, 80)
except SaldoInsuficienteError as e:
    print(e)

print(alicia)  # Saldo esperado: 100 + 80 = 180 si se pudo. En este caso no se pudo. Por tanto sigue en 100.
print(bob)     # Saldo esperado: 70 - 80 = no suficiente, se lanzará excepción

# 4. Retirar 50 unidades de saldo a Alicia
try:
    alicia.retirar_dinero(50)
except SaldoInsuficienteError as e:
    print(e)

print(alicia)  # Saldo esperado: 180 - 50 = 130
```

```python
# 37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: reemplazar_palabras , procesar_texto , contar_palabras y/o eliminar_palabra.
# Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función. Código a seguir:
# 37.1 Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
# 37.2 Crear una función reemplazar_palabras para remplazar palabra_original del texto por una palabra_nueva.
# 37.3 Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto sin la palabra eliminada.
# 37.4 Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.
# Caso de uso:Comprueba el funcionamiento completo de la función procesar_texto
```

```python
from collections import Counter

# ------------------- Funciones auxiliares -------------------

def contar_palabras(texto):
    """
    Cuenta el número de veces que aparece cada palabra en el texto.

    Args:
        texto (str): Texto a analizar.

    Returns:
        dict: Diccionario con palabras como clave y conteo como valor.
    """
    palabras = texto.split()
    return dict(Counter(palabras))


def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """
    Reemplaza todas las apariciones de una palabra por otra en el texto.

    Args:
        texto (str): Texto original.
        palabra_original (str): Palabra a reemplazar.
        palabra_nueva (str): Palabra nueva.

    Returns:
        str: Texto con palabras reemplazadas.
    """
    return texto.replace(palabra_original, palabra_nueva)


def eliminar_palabra(texto, palabra_a_eliminar):
    """
    Elimina todas las apariciones de una palabra en el texto.

    Args:
        texto (str): Texto original.
        palabra_a_eliminar (str): Palabra a eliminar.

    Returns:
        str: Texto con la palabra eliminada.
    """
    palabras = texto.split()
    palabras_filtradas = [palabra for palabra in palabras if palabra != palabra_a_eliminar]
    return ' '.join(palabras_filtradas)


# ------------------- Función principal -------------------

def procesar_texto(texto, opcion, *args):
    """
    Procesa un texto según la opción indicada.

    Args:
        texto (str): Texto a procesar.
        opcion (str): Una de las opciones: 'contar', 'reemplazar', 'eliminar'.
        *args: Argumentos variables según la opción.

    Returns:
        str|dict: Resultado del procesamiento.
    """
    opcion = opcion.lower()

    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        if len(args) != 2:
            raise ValueError("Para 'reemplazar' se requieren 2 argumentos: palabra_original, palabra_nueva")
        return reemplazar_palabras(texto, args[0], args[1])
    elif opcion == "eliminar":
        if len(args) != 1:
            raise ValueError("Para 'eliminar' se requiere 1 argumento: palabra_a_eliminar")
        return eliminar_palabra(texto, args[0])
    else:
        raise ValueError("Opción no válida. Debe ser 'contar', 'reemplazar' o 'eliminar'.")


# ------------------- Caso de uso PROCESAR_TEXTO -------------------

texto_original = "hola mundo hola estudiantes, bienvenidos al entorno python"
print(texto_original)

# 1. Contar palabras
conteo = procesar_texto(texto, "contar")
print("Conteo de palabras:", conteo)

# 2. Reemplazar palabras
texto_reemplazado = procesar_texto(texto, "reemplazar", "hola", "XYZ")
print("Texto reemplazado:", texto_reemplazado)

# 3. Eliminar palabra
texto_eliminado = procesar_texto(texto, "eliminar", "python")
print("Texto con palabra eliminada:", texto_eliminado)

```

```python
# Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo", "circulo" o "triangulo" ) y 
# 'datos' (una tupla con los datos necesarios para calcular el área de la figura).
```

```python
import math

def calcular_area(figura, datos):
    """
    Calcula el área de una figura geométrica.

    Args:
        figura (str): Tipo de figura ('triangulo', 'rectangulo', 'circulo').
        datos (tuple): Datos necesarios para calcular el área:
            - triangulo: (base, altura)
            - rectangulo: (largo, ancho)
            - circulo: (radio,)

    Returns:
        float: Área de la figura.
    """
    figura = figura.lower()

    if figura == "triangulo":
        if len(datos) != 2:
            raise ValueError("Para un triángulo se requieren dos datos: base y altura")
        base, altura = datos
        return (base * altura) / 2
    elif figura == "rectangulo":
        if len(datos) != 2:
            raise ValueError("Para un rectángulo se requieren dos datos: largo y ancho")
        largo, ancho = datos
        return largo * ancho
    elif figura == "circulo":
        if len(datos) != 1:
            raise ValueError("Para un círculo se requiere un dato: radio")
        radio = datos[0]
        return math.pi * radio ** 2
    else:
        raise ValueError("Figura no válida. Debe ser 'triangulo', 'rectangulo' o 'circulo'")


# ------------------- Ejemplos de uso -------------------

print("Área del triángulo:", calcular_area("triangulo", (5, 10)))   # 25.0
print("Área del rectángulo:", calcular_area("rectangulo", (8, 3)))  # 24
print("Área del círculo:", calcular_area("circulo", (4,)))          # 50.26548245743669
```

```python
def calcular_precio_final():
    """
    Calcula el precio final de una compra aplicando un cupón de descuento si el usuario lo tiene.
    """
    try:
        # 41.1. Solicitar el precio original
        precio_original = float(input("Introduce el precio original del artículo (€): "))
        if precio_original < 0:
            print("Error: El precio no puede ser negativo.")
            return

        # 41.2. Preguntar si tiene cupón de descuento
        tiene_cupon = input("¿Tienes un cupón de descuento? (sí/no): ").strip().lower()

        precio_final = precio_original

        if tiene_cupon == "sí" or tiene_cupon == "si":
            # 41.3. Solicitar valor del cupón
            try:
                valor_cupon = float(input("Introduce el valor del cupón (€): "))
                # 41.4. Aplicar descuento si es válido
                if valor_cupon > 0:
                    precio_final -= valor_cupon
                    if precio_final < 0:
                        precio_final = 0  # Evitar precio negativo
                else:
                    print("El valor del cupón debe ser mayor a cero. No se aplica descuento.")
            except ValueError:
                print("Error: Debes introducir un número válido para el cupón.")

        elif tiene_cupon == "no":
            print("No se aplica descuento.")
        else:
            print("Respuesta no válida. No se aplica descuento.")

        # 41.5. Mostrar precio final
        print(f"El precio final de la compra es: {precio_final:.2f} €")

    except ValueError:
        print("Error: Debes introducir un número válido para el precio.")


# ------------------- Ejecutar programa -------------------
if __name__ == "__main__":
    calcular_precio_final()

# Ejemplo de uso:
# Input de 'Introduce el precio original del artículo (€):' 25
# Input de '¿Tienes un cupón de descuento? (sí/no):' si
# Input de 'Introduce el valor del cupón (€):' 3
# Output: 'El precio final de la compra es: 22.00 €'
```
